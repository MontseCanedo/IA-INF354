```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("../drinks.csv")
data
```

Como vemos, los datos son muy dispersos y desordenados, se necesita darle estructura para poder trabajar con ellos.

```python
data1 = data[data["total_litres_of_pure_alcohol"] > 0.0]
data1
```

aquí hemos creado un nuevo dataFrame a partir del original, donde eliminamos las filas de datos con consumo de alcohol igual a 0, para tener puros datos con valores significativos.

```python
data1.describe(include="all")
```

```python
data2 = data1.iloc[: , 1:5]
x = data2.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
data2 = pd.DataFrame(x_scaled)
```
